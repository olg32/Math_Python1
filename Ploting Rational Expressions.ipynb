```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(-3,4)
y = np.zeros(len(x))

for i in range(0,len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x,y,'s-')
plt.show()
```

```python
x = np.linspace(-3,3,20)
y = 2 + np.sqrt(abs(x))

# magenta squares lines ms-
plt.plot(x,y,'ms-')
plt.show()
```
