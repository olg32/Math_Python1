```python
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5
plt.plot(x,y,'ro')
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid()
plt.show()
```

```python
x = [-4,5,2,4,5,6,3]
y = [0,-3,8,3,2,4,-5]

for i in range(0,len(x)):
    plt.plot(x[i],y[i],'o', label='point %s' %(1+i))

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```
