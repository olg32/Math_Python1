```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#  y = mx + b

x = [-5,5]
m = [.7,-1.25]
b = [-2,.75]

for i in range(0,len(m)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x,y,label='y = %sx + %s' %(m[i],b[i]))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')
    
plt.legend()
plt.show()
```
